```python
import numpy as np
list1=[1,2,3]
array1=np.array(list1)
```

Mathematical Operations using Numpy

```python
list2=[3,2,1]
array2=np.array(list2)
```

```python
print(f'addition of two arrays:,{array1+array2}')
print(f'multiplication of two arrays",{array1*array2}')
print(f'power of two arrays:,{array1**array2}')
```

Mathematical Function in Numpy

```python
print(f'sin value of array1:,{np.sin(array1)}')
print(f'Log2 value of array1:,{np.log2(array1)}')
print(f'Exponential value of array1:,{np.exp(array1)}')
```

```python
people=['Ann','Brandon','Chen','David']
age=[21,12,32,45]
weight=[55,37,77,68]
height=[160,135,170,165]
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(age,weight,height)
plt.title("Weight vs Height:",fontsize=16)
plt.show()
```

```python
import numpy as np
plt.figure(figsize=(7,5))
plt.hist(weight,color='aqua',edgecolor='k',alpha=0.75,bins=5)
plt.title("Histogram of patient weight",fontsize=15)
plt.xlabel("People",fontsize=15)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.show()
```

```python
plt.figure(figsize=(4,4))
plt.title("People's weight in kgs",fontsize=16)
plt.bar(x=people,height=weight,width=0.6,color="orange",edgecolor="b",alpha=0.6)
plt.xlabel("people",fontsize=15)
plt.xticks(fontsize=14,rotation=30)
plt.yticks(fontsize=14)
plt.ylabel("weigt(in kgs)", fontsize=16)
plt.show()
```

```python
plt.figure(figsize=(4,4))

plt.hist(weight,color="yellow",edgecolor="k",alpha=0.6,bins=9)
plt.title("Histogram of people's Weight:",fontsize=16)
plt.xlabel("Weight(in Kgs)")
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.show()
```
